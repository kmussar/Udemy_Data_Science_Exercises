Note - See QA section. Many alternative solutions to this. Jose's solution doesn't take into account punctuation in their titles. (i.e. Chief, Justice).

```python
#Another approach - Jose's appoach 
# function to check each title - lowercase it and check if it contains 'chief'
def chief_string(title):
    if 'chief' in title.lower().split(): 
        return True
    else: 
        return False 
# run this function on the whole table 
# use sum to add up all the true values to get the total number of title with chief in them
sum(sal['JobTitle'].apply(lambda x: chief_string(x)))

"""This is actually problematic because it uses split which creates some strings 
containing commas. """
```

```python
# Jose's answer
```

```python
# Better answer - takes into account that some employees are there for multiple years 
# and thus have repeated entries in the dataset. 

import numpy as np

def cleanName(name):
    #remove punctuation, there's probably a better way to do this. This is from https://stackoverflow.com/a/3939381/1300676
    for char in name:
        if char in " ?.!/;:":
            name = name.replace(char,'')
    #Remove white space and capitalization
    names = np.array(name.lower().strip().split())
    #Remove initials
    lengths = [len(str) > 1 for str in names]
    
    return ' '.join(names[lengths])
 
#Filter on job titles that contain chief
salChiefs = sal[sal["JobTitle"].apply(lambda title: "chief" in title.lower())].copy()
#Clean employee names
salChiefs["EmployeeName"] = salChiefs["EmployeeName"].apply(cleanName)
#Final count
len(salChiefs.groupby("EmployeeName").count().index)

```
